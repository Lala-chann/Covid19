```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
file = pd.read_csv('compact.csv')
file.head()
```

```python
print(file.columns.tolist())
```

```python
selected_columns = file[['country','date','hosp_patients','weekly_hosp_admissions','total_vaccinations','people_fully_vaccinated']]
selected_columns.head()
```



>  Data about ***Azerbaijan***



```python
az_data = selected_columns[selected_columns['country']== 'Azerbaijan']
az_data_index = az_data.set_index('date')
az_data_index.head()
```

```python
az_data_index.index = pd.to_datetime(az_data_index.index)
az_data_filtered = az_data_index.loc['2022-01-01':'2022-01-05']
display(az_data_filtered)
```



> Data about ***Turkey***



```python
tr_data = selected_columns[selected_columns['country'] == 'Turkey']
tr_index_data = tr_data.set_index('date')
tr_index_data.head()
```

```python
tr_index_data.index = pd.to_datetime(tr_index_data.index)
tr_data_filtered = tr_index_data.loc['2022-01-01':'2022-01-05']
display(tr_data_filtered)
```



> Data about ***Russia***



```python
rus_data = selected_columns[selected_columns['country'] == 'Russia']
rus_data_index = rus_data.set_index('date')
rus_data_index.head()
```

```python
rus_data_index.index = pd.to_datetime(rus_data_index.index)
rus_data_filtered = rus_data_index.loc['2022-01-01':'2022-01-05']
display(rus_data_filtered)
```



> Data about ***Georgia***



```python
ge_data = selected_columns[selected_columns['country']== 'Georgia']
ge_data_index = ge_data.set_index('date')
ge_data_index.head()
```

```python
ge_data_index.index = pd.to_datetime(ge_data_index.index)
ge_data_filtered = ge_data_index.loc['2022-01-01':'2022-01-05']
display(ge_data_filtered)
```



> **Total Vaccination** for each country





```python
az_total_vac = az_data_filtered['total_vaccinations'].sum()
tr_total_vac = tr_data_filtered['total_vaccinations'].sum()
rus_total_vac = rus_data_filtered['total_vaccinations'].sum()
ge_total_vac = ge_data_filtered['total_vaccinations'].sum()
```

```python
float_az = az_total_vac
r_az = round(float_az / 1_000_000, 1)
print(r_az)

float_tr = tr_total_vac
r_tr = round(float_tr / 1_000_000,1)
print(r_tr)

float_ru = rus_total_vac
r_ru = round(float_ru / 1_000_000,1)
print(r_ru)

float_ge = ge_total_vac
r_ge = round(float_ge / 1_000_000,1)
print(r_ge)
```

```python
data = {'country': ['Azerbaijan', 'Turkey','Russia','Georgia'],'total_vaccination':[r_az,r_tr,r_ru,r_ge]}
df = pd.DataFrame(data)
df.head()
```



> Visualization of data using **a histogram and a pie chart**



```python
colors = ['#A30000','#470000','#750000','#D10000',]

plt.figure(figsize=(6,6))
plt.bar(df['country'],df['total_vaccination'],color = colors)
plt.xlabel('Country')
plt.ylabel('Total Vaccinations')
plt.title('Vaccination Distribution Across Countries')
plt.yticks([50,100,200,300,400,500])
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.show()
```



> Example : **Named Colors**



```python
data_pie = np.array([r_az,r_tr,r_ru,r_ge])
labels = ['Azerbaijan','Turkey','Russia','Georgia']
colors = ['#FF8A8A','#750000','#D10000','#FFE6E6',]
plt.pie(data_pie,labels=labels,colors=colors,autopct='%1.1f%%')
plt.title('Vaccination Distribution Across Countries')
plt.axis('equal')
plt.show()
```



> Example : using a **colormap**


```python
data_pie_chart = pd.Series([r_az,r_tr,r_ru,r_ge])
labels = ['Azerbaijan','Turkey','Russia','Georgia']
plt.pie(data_pie_chart,labels=labels,colors=plt.cm.Paired.colors,autopct='%1.1f%%')
plt.axis('equal')
plt.title('Vaccination Distribution Across Countries')
plt.show()
```
